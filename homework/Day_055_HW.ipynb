```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn import datasets

np.random.seed(5)

%matplotlib inline
```

```python

iris = datasets.make_blobs(centers=3, n_features=4)
X = iris[0]
y = iris[1]
```

```python

estimators = [('k_means_8', KMeans(n_clusters=8)), #要进行的分类的个数，即上文中k值，默认是8
              ('k_means_3', KMeans(n_clusters=3)),
              ('k_means_bad_init', KMeans(n_clusters=3, n_init=1,
                                               init='random'))]
```

```python

fignum = 1
titles = ['8 clusters', '3 clusters', '3 clusters, bad initialization']
for name, est in estimators: #第一個陣列名字 ，第二陣列參數
    fig = plt.figure(fignum, figsize=(8, 6))#plt size
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134) #視角
    
    ## fit data
    est.fit(X)
    
    labels = est.labels_ #設置標籤

    ax.scatter(X[:, 3], X[:, 0], X[:, 2], #調整具有代表性的三維空間
               c=labels.astype(np.float), edgecolor='k') #C顏色  edgecolor邊緣顏色

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])#設置XYZ
    ax.set_title(titles[fignum - 1])#名稱
    ax.dist = 12 #距離
    fignum = fignum + 1

# Plot the ground truth
fig = plt.figure(fignum, figsize=(8, 6))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

for name, label in [('cls0', 0),('cls1', 1),('cls2', 2)]:
    ax.text3D(X[y == label, 3].mean(), #xyz
              X[y == label, 0].mean(),
              X[y == label, 2].mean() + 2, name,
              horizontalalignment='center', #水平線性
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))#顏色
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)  #按照序号a对choices中的数进行选择  默认的是raise
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=y, edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_title('Ground Truth')
ax.dist = 12

#fig.show()
```
